Side by side barplot()

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (20, 32, 34, 20, 27)

ind = np.arange(N)
width = 0.35

#이렇게 하면 한 화면에 그래프 2개 그릴 수 있다.
# 1개씩 그리고 싶으면 p1,p2다음에 각각 plt.show()를 집어 넣어주면 된다.
p1 = plt.bar(ind-width/2, menMeans, width, color ='#d62728')
# plt.show()
p2 = plt.bar(ind+width/2, womenMeans, width)
# plt.show()

plt.ylabel('No test')
plt.xlabel('ChainSmokers')
plt.xticks(ind, ('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0,81,10))
print(p1[0])
plt.legend((p1[0],p2[0]), ('Men','Women')) #p1[0]. p2[0] 은 크기지정으로 사료됨
```

Stacked barplot

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (50, 32, 34, 20, 27)

ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, menMeans, width, color ='#d62728')
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans)

plt.ylabel('No test')
plt.xlabel('ChainSmokers')
plt.xticks(ind, ('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0,81,10))
print(p1[0])
plt.legend((p1[0],p2[0]), ('Men','Women')) # p1[0]. p2[0] 은 크기지정으로 사료됨
```

### <font color = white>Matplotlib module : boxplot()

```python
import matplotlib.pyplot as plt
import numpy as np

spread = np.random.rand(50)*100 # 0~100 범위 임의의 수 50개,
center = np.ones(25) * 50 # 숫자50(1에 *50했으니깐) 25개
print(center)
flier_high = np.random.rand(10) * 100 + 100 # 100~200범위 임의의 수 10개
flier_low = np.random.rand(10) * -100 # -100~0 범위 임의의 수 10개 이상치
data = np.concatenate((spread, center, flier_high, flier_low),0) # 0 있으면 그냥 concatenate해주는거
print(data)
# 1 있으면 [3,4,7]
#          [5,6,8]

plt.boxplot(data) # plt.boxplot에 option없이 data만 집어넣음.
```

### <font color = white>Matplotlib module : Multiple Plots with subplot()
- figure 그리고 subplot 위치정해줘서 그려주는 것은 subplot함수를 쓴다 / subplot(위치정보) <br>/ subplot(211) 2행1열의 1번째

```python
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211) # 2행 1열의 1번 subwindow냐
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k') # 여기는 그래프 2개 그려준거

# plt.figure(2) : 이게있으면 2번째 캔버스에 그려준다.
plt.subplot(212) # 2행 1열의 2번째 subwindow냐 figure
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```
